```python
import matplotlib.pyplot as plt
%matplotlib inline

# Data storages 
import numpy as np
import pandas as pd

# helpers
from IPython.display import display
```

## Color codes
* b: blue.
* g: green.
* r: red.
* c: cyan.
* m: magenta.
* y: yellow.
* k: black.
* w: white.

# 1. Plotting using methods on `AxesSubplot`

```python
samples = np.round(np.random.randn(4,6), 2)
display(samples)
```

```python
# ax is an ndarray of size (nrows, ncols) containing handle to AxesSubplot objects
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,10))

top_left_axis = ax[0,0]
# o-> circle; s-> square; ^-> triangle
top_left_axis.plot(samples, 's')

top_left_axis.set_title('Axis Based Plot')
top_left_axis.set_xlabel('X-axis is Rows')
top_left_axis.set_ylabel('Y-axis is Columns')

# a list of [xmin, xmax, ymin, ymax]
top_left_axis.axis([-1, samples.shape[0]+1, -1, samples.shape[1]+1])
```

# 2. Plotting using methods on `AxesSubplot` and Pandas `DataFrame`

```python
samples = pd.DataFrame(np.round(np.random.randn(4,6), 2))
display(samples)
```

```python
# ax is an ndarray of size (nrows, ncols) containing handle to AxesSubplot objects
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,10))

top_left_axis = ax[0,0]
samples.plot(ax=top_left_axis, kind='bar')

top_left_axis.set_title('Axis Based Plot')
top_left_axis.set_xlabel('X-axis is Rows')
top_left_axis.set_ylabel('Y-axis is Columns')
top_left_axis.set_xticklabels(["Row 1", "Row 2", "Row 3", "Row 4"], rotation=0)

# a list of [xmin, xmax, ymin, ymax]
top_left_axis.axis([-1, samples.shape[0]+1, samples.values.min()-0.5, samples.values.max()+0.5])

top_left_axis.text(-0.6, top_left_axis.get_ylim()[1]+0.05, "My Fav")
```

# 3. Making a scatter plot

```python
samples = pd.DataFrame(np.round(np.random.randn(10,2), 2), columns=['Bert', 'Ernie'])
display(samples.head())
```

```python
fig, ax = plt.subplots(figsize = (5,5))
bert_scores = samples.loc[:,'Bert']
ernie_scores = samples.loc[:,'Ernie']
# s -> represents size of the marker 
ax.scatter(x=bert_scores,
           y=ernie_scores,
           marker='*', s=1000, edgecolors='r', facecolors='b')
ax.set_title('Bert Vs. Ernie')
ax.set_xlabel('Bert')
ax.set_ylabel('Ernie')

# 
ax.arrow(0, 0, -1, 0, head_width=.1, head_length=0.1, linewidth=2, color='k')
ax.text(-1.6, 0, "Arrow")

# a list of [xmin, xmax, ymin, ymax]
top_left_axis.axis([bert_scores.min()-.5, bert_scores.max()+0.5, ernie_scores.min()-0.5, ernie_scores.max()+0.5])
```
