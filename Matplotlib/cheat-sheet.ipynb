```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import display
```

# 1. Plotting using methods on `AxesSubplot`

```python
samples = np.round(np.random.randn(4,6), 2)
display(samples)
```

```python
# ax is an ndarray of size (nrows, ncols) containing handle to AxesSubplot objects
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,10))

top_left_axis = ax[0,0]
# o-> circle; s-> square; ^-> triangle
top_left_axis.plot(samples, 's')

top_left_axis.set_title('Axis Based Plot')
top_left_axis.set_xlabel('X-axis is Rows')
top_left_axis.set_ylabel('Y-axis is Columns')

# a list of [xmin, xmax, ymin, ymax]
top_left_axis.axis([-1, samples.shape[0]+1, -1, samples.shape[1]+1])
```
